```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
x = sonar.drop('Y',1, inplace=False)
y = sonar['Y']
```

```python
model1 = DecisionTreeClassifier()
model2 = KNeighborsClassifier()
model3 = ExtraTreesClassifier()
model4 = SVC(kernel='rbf')
model5 = SVC(kernel='poly')
model6 = SVC(kernel='linear')
model7 = SVC(kernel='sigmoid')
model8 = LogisticRegression()

model = [model1, model2, model3, model4, model5, model6, model7, model8]
model_names = ['DecisionTree Model', 'KNN Model', 'ExtraTrees Model', 'SVC rbf Model',
              'SVC poly Model', 'SVC linear Model', 'SVC sigmoid Model', 'Logistic Regression Model']

def classify(model, modelname, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.3, random_state=101)
    # Model training
    model.fit(x_train, y_train)
    print("Accuracy score for {} is:".format(modelname),model.score(x_train, y_train))
    #Cross validation
    score = cross_val_score(model, x,y, cv=5)
    print("Cross validation score for {} is:".format(modelname),np.mean(score))
    return np.mean(score)
```

```python
c_scores = []
for i,j in enumerate(model):
    c_scores.append(classify(j, model_names[i], x, y))
```

```python
print("Best model for wine is:",model_names[c_scores.index(max(c_scores))])
```
